#### Задание 1 (10 баллов)

Необходимо реализовать одномерный [бинарный клеточный автомат](http://mathworld.wolfram.com/ElementaryCellularAutomaton.html). 
Продемонстрировать работу программы на правилах 158 и 18. Сделать не менее 100 итераций. За использование каждого цикла `for` более 1 раза - **штраф** 5 баллов.

````python
# Перевод десятичного числа (номера правила) в двоичную систему счисления
def translate_rule(rule):
  if rule > 255 or rule < 0:
    raise Exception("ERROR. Rule must be in range [0,255]")
  configuration = bin(rule)[2:].zfill(8)[::-1]
  return configuration

# Функция перехода из старого состояния в новое
def get_new_condition(configuration, input):
  output = []
  n = len(input)
  # Единственный цикл For
  for i in range(0, n):
    code = input[(i-1)%n]*4 + input[i]*2 + input[(i+1)%n]
    output.append(int(configuration[code]))
  return output


# Рекурсивная функция для пробега всех итераций
def repeat_iterations(input, configuration, iterations):
  print(''.join(map(str, input)))
  new_condition = get_new_condition(configuration, input)
  if iterations > 0:
    repeat_iterations(new_condition, configuration, iterations - 1)

# Вспомогательная функция для запуска процесса расчёта состояний
def start_process(input, rule, iterations):
  configuration = translate_rule(rule)
  repeat_iterations(input, configuration, iterations)

# Пример из задания
rule = 158
input = [0]*100 + [1] + [0]*100
print("Rule", rule)
start_process(input,rule,100)

# Пример из задания
rule = 18
input = [0]*100 + [1] + [0]*100
print("Rule", rule)
start_process(input,rule,201)

# Просто красивые треугольники
rule = 110
input = [0]*100 + [1]
print("Rule", rule)
start_process(input,rule,100)
````

#### Задание 2 (20 баллов)

В списке vecs - координаты `x`, `y` векторов в двумерном пространстве.
Рассчитать:
1. Длину ломаной линии, соединяющей последовательно концы векторов;
2. Углы между предыдущим и следующим векторами;
3. Площади треугольников, вершины которых заданы концами i, i+1, i+2 векторов.

Использование любых циклов, кроме встроенных в `Numpy` **запрещено**, за каждый такой цикл - **штраф** 5 баллов.

````python
import numpy as np

vecs = [[6.44, 5.05],
        [3.59, 7.71],
        [8.81, 2.02],
        [8.71, 7.8 ],
        [5.51, 2.53],
        [6.35, 4.89],
        [7.88, 3.52],
        [4.89, 5.83],
        [1.66, 5.24],
        [3.1 , 8.91],
        [5.11, 8.22],
        [7.06, 7.15],
        [2.15, 2.78],
        [7.25, 8.94],
        [1.62, 4.97],
        [8.86, 4.11],
        [8.59, 8.38],
        [2.18, 7.01],
        [8.24, 8.57],
        [9.97, 1.82]]

vectors = np.array(vecs)

# [x_(i+1)-x_(i),y_(i+1)-y_(i)]
difference = vectors[1:] - vectors[:-1]

# 1
len_vecs = np.linalg.norm(difference, axis = 1)
sum_len_vecs = len_vecs.sum()
print("Длину ломаной линии, соединяющей последовательно концы векторов:", sum_len_vecs)

# 2
scalar = np.sum(difference[1:] * difference[:-1], axis = 1)
len_ab = len_vecs[1:]*len_vecs[:-1]
fi = np.arccos(scalar/len_ab)
print("Углы между предыдущим и следующим векторами:", fi, sep="\n")

# 3
S = len_vecs[1:]*len_vecs[:-1]*np.sin(fi)
print("Площади треугольников, вершины которых заданы концами i, i+1, i+2 векторов:", S, sep="\n")
````
