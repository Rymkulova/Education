```python
phd_degree = len(df[df['ind-degree'] == 'phd'])
bachelor_degree = len(df[df['ind-degree'] == 'bachelor'])
master_degree = len(df[df['ind-degree'] == 'master'])

males = len(df[df['gender'] == 'male'])
females = len(df[df['gender'] == 'female'])
other = len(df[df['gender'] == 'other'])
```
